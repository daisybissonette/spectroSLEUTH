## SpectroSLEUTH Tutorial Notebook
Here is a quick tutorial of how to import and use `spectroSLEUTH` to find emission lines in a spectrum. The example spectra used here are SDSS spectra included the test suite for `spectroSLEUTH`. 

The testing file redshift are as follows:
Test 1: 0.06456,
Test 2: 0.08428,
Test 3: 0.02067

```python
# import the main script in spectroSLEUTH, "sleuth" 
import spectroSLEUTH.sleuth as sleuth

# create an instance of the emission_lines object class, with a known redshift and user-chosen prominence
el = sleuth.emission_lines('tests/test_spectrum_1.fits', 0.06456, prominence=30)

# run the line_identification function to see the emission lines plot and table
matches, fig = el.line_identification()
fig.savefig('test_identify.png', bbox_inches ="tight", pad_inches = 0.05, dpi=300)
```
